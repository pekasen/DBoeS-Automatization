```python
kreisnamen_path = 'data/btw17_wahlkreisnamen.csv'
candidates_path = 'data/Kandidierende21Zwischenstand.csv'
old_candidates_path = '../db/reviewed/BT-Wahl 2021.csv'
editor_name = 'fvm'
```

```python
import pandas as pd
```

```python
from datetime import datetime
import time
```

```python
df_kreisnamen = pd.read_csv(kreisnamen_path, sep=';', dtype='str')
df_kreisnamen
```

```python
df_candidates = pd.read_csv(candidates_path, sep=';', dtype='str')
df_candidates
```

```python
df_old_candidates = pd.read_csv(old_candidates_path, dtype='str')
df_old_candidates
```

```python
def convert_district_names(df):
    district_names = df_kreisnamen[['WKR_NR','WKR_NAME']]
    
    district_names['WKR_NR'] = district_names['WKR_NR'].astype(str)
    
    df = df.merge(district_names, left_on='Wahlkreis', right_on='WKR_NR', how='left')
    
    df['Wahlkreis'] = df['WKR_NAME']
    
    del df['WKR_NAME']
    del df['WKR_NR']
    
    df['modified_by'] = editor_name
    df['modified_at'] = str(datetime.now())
    
    result = df
    
    return result
    
candidates_with_district_names = convert_district_names(df_candidates)
candidates_with_district_names
```

```python
df_old_candidates.merge(candidates_with_district_names, how='outer', on=['Name', 'Partei', 'Wahlkreis'], suffixes=(None,"_new"))
```
