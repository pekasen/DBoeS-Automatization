```python
kreisnamen_path = 'data/btw17_wahlkreisnamen.csv'
candidates_path = 'data/Kandidierende21Zwischenstand.csv'
old_candidates_path = '../db/reviewed/BT-Wahl 2021.csv'
editor_name = 'fvm'
```

```python
import pandas as pd
```

```python
from datetime import datetime
```

```python
df_candidates = pd.read_csv(candidates_path, sep=';', dtype='str')
df_candidates
```

```python
df_old_candidates = pd.read_csv(old_candidates_path, dtype='str')
df_old_candidates
```

```python
import uuid

def merge_candidates(on):
    
    result = df_old_candidates.merge(df_candidates, how='outer', on=on, suffixes=(None,"_new"))
    
    result['modified_by_new'] = editor_name
    result['modified_at'] = datetime.now()
    
    for column in result.columns:
        if column[-4:] == '_new' and column != 'id_new':
            result[column[:-4]][result[column].notna()] = result[column][result[column].notna()]
            del result [column]
    
    del result['id_new']
    
    number_of_new_entries = len(result['id'][result['id'].isna()])
    
    result['created_at'][result['id'].isna()] = str(datetime.now())
    result['created_by'][result['id'].isna()] = editor_name
    
    result['id'][result['id'].isna()] = [uuid.uuid4() for _ in range(number_of_new_entries)]
    
    return result
    
df = merge_candidates(['Name', 'Partei', 'Wahlkreis'])

df.to_csv('data/BT-Wahl 2021.csv', index=False)
```
